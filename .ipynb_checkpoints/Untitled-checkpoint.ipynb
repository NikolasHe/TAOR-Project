```python
import pandas as pd

#import open source data
df = pd.read_csv("vehicles.csv", low_memory=False)
df.head()

# examine car manufacturers included in dataframe
df['make'].unique()
```

```python
# filter only data containing ford vehicles (as Shelby are modified Ford vehicles I included them in the dataset)
df_ford = df.query('make in ["Ford", "Shelby"]')
print('Number of entries:',len(df_ford))
print('Number of different car models:',len(df_ford['model'].unique()))
print(df_ford['VClass'].unique())
```

```python
list(df_ford.columns) 

df_ford_short = df_ford[['barrels08',
 'barrelsA08',
 'city08',
 'city08U',
 'cityA08',
 'cityE',
 'cityUF',
 'co2',
 'co2A',
 'comb08',
 'combA08',
 'combE',
 'cylinders',
 'displ',
 'drive',
 'feScore',
 'fuelCost08',
 'fuelCostA08',
 'fuelType',
 'fuelType1',
 'highway08',
 'highwayA08',
 'highwayCD',
 'highwayE',
 'highwayUF',
 'hlv',
 'hpv',
 'lv2',
 'lv4',
 'make',
 'model',
 'mpgData',
 'range',
 'rangeCity',
 'rangeCityA',
 'rangeHwy',
 'rangeHwyA',
 'VClass',
 'year',
 'youSaveSpend',
 'atvType',
 'fuelType2',
 'rangeA',
 'evMotor',]]
```

```python
#piechart for distribution of number of cylinders
df_ford_short.groupby(['cylinders']).count().plot(kind='pie',y='model')
#piechart for distribution of vehicle class
df_ford_short.groupby(['VClass']).count().plot(kind='pie',y='model',legend=False)
```
